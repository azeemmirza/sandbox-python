#### **Task 01:**

- Get user input using input('Enter your age: ').
- If user is 18 or older, give feedback: You are old enough to drive.
- If below 18 give feedback to wait for the missing amount of years.

```python
try:
  age = int(input('Enter your age: '))
  
  if age > 18:
    print('You are old enough to drive.')
  else:
    years_left = 18 - age
    print(f'You are missing { years_left } of years.')
except ValueError:
  print('[Error]: Value should be valid integer')
```

#### **Task 02:**

- Compare the values of my_age and your_age using if … else.
- Who is older (me or you)? Use input(“Enter your age: ”) to get the age as input.
- You can use a nested condition to print 'year' for 1 year difference in age, 'years' for bigger differences, and a custom text if my_age = your_age.

```python
try:
  person_a_age = int(input('Enter First person\'s age: '))
  person_b_age = int(input('Enter Second person\'s age: '))

  if person_a_age > person_b_age:
    diff = person_a_age - person_b_age
    print(f'Person A is { diff } year(s) older than Person B')
  
  elif person_b_age > person_a_age:
    diff = person_b_age - person_a_age
    print(f'Person B is { diff } year(s) older than Person A')
  
  else:
    print('Person A is same as Person B')

except ValueError:
  print('[Error]: Value should be valid integer')
```

#### **Task 03:**

- Get two numbers from the user using input prompt.
- If a is greater than b return a is greater than b.
- If a is less b return a is smaller than b, else a is equal to b.

```python
try: 
  a = int(input('Enter value for a: '))
  b = int(input('Enter value for b: '))

  if a > b:
    print('a is greater than b')
  
  elif a < b:
    print('a is smaller than b')
  
  else:
    print('a is euqal to b')

except ValueError:
  print('[Error]: Value should be valid integer')
```
